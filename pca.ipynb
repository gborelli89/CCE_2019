# PRINCIPAL COMPONENTS ANALYSIS (PCA)

Usado para reduzir número de variáveis de um problema. A partir de um número de observações com variáveis supostamente correlacionadas, calcula os componentes principais (problema de autovalor e autovetor ou SVD).

Inicialmente as variáveis são escaladas (subtrai-se a média e divide-se pelo desvio padrão). É possível chegar em um problema de autovalor e autovetor para a matrix de covariância.

Pode ser utilizada em conjunto com algum algoritmo de *clustering* para problemas de classificação (*machine learning*). 

É útil para compreender problemas com múltiplas variáveis e na formulação de modelos.

É relativamente simples de ser feito em R. Vamos importar a base de dados *iris* .

OBS.: Exemplo dado no video https://youtu.be/xKl4LJAXnEA

```r
data(iris)
head(iris)
summary(iris)
```

A função **prcomp** calcula os componentes principais. Usar *scale = TRUE*

```r
pca <- prcomp(iris[,-5], scale=TRUE, center=TRUE)
pca
```

```r
summary(pca)
```

```r
# Utilizado apenas para reduzir tamanho dos gráficos
require(repr)
options(repr.plot.width=7, repr.plot.height=4)

barplot(pca$sdev^2/sum(pca$sdev^2)*100, main = "Proporção da variância",names.arg = paste0("PC",1:4))
```

```r
head(pca$x, n=7)
```

Vamos utilizar os dois primeiros componentes principais, responsáveis por aproximadamente 96 % da variância dos dados. A biblioteca ggplot será utilizada para plotar.

```r
iris2 <- cbind(iris, pca$x[,1:2])
head(iris2)
```

## Exemplo de aprendizagem na supervisionada utilizando K-means após PCA

K-means é um algoritmo simples que pode ser utilizado para encontrar clusters de forma não supervisonada, isto é, quando não temos um grupo que conhecemos a classificação para treinamento. No exemplo com a base de dados *iris* isto equivaleria a não ter a coluna de espécies.

O algoritmo pode ser utilizado com várias variáveis, pode ser interessante reduzir o número de variáveis utilizando, por exemplo, o PCA. Normalmente um problema com muitas variáveis tende a necessitar de um número maior de dados. Para a base de dados *iris* a utilização do PCA não faz muita diferença.

Funcionamento do algoritmo:

Inicialmente, dado o número de *nc* de clusters, são gerados *nc* grupos de dados de forma aleatória. A média Euclidiana de cada grupo pode ser calculada. Cada ponto é então classificado novamente de acordo com a sua distância para o centro dos clusters. As interações ocorrem até que os centros dos clusters permanecem os mesmos, dado uma tolerância pré determinada.

Para o exemplo vamos utilizar os resultados obtidos com PCA

```r
# modelo com 2 clusters
m2c <- kmeans(iris2[,6:7], centers = 2)

m2c
```

```r
plot(iris2[,6:7], col= iris2$Species, pch=m2c$cluster)
```

```r
# modelo com 3 clusters
m3c <- kmeans(iris2[,6:7], centers = 3)

m3c
```

```r
plot(iris2[,6:7], col= iris2$Species, pch=m3c$cluster)
```

```r
funSS <- function(nc){
    ss <- c()
    for(i in nc){
        m <- kmeans(iris2[,6:7], centers = i)
        ss <- c(ss, m$betweenss/m$totss)
    }
    return(ss*100)
}

k <- 1:7
plot(k, funSS(k), t='b', ylab='betweenSS/totSS', xlab='clusters')
```
