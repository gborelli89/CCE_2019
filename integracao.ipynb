## Regra do trapézio

A regra do trapézio faz uso da interpolação linear entre os pontos. Pode-se obter a expressão abaixo.

$$ \int_a^b f(x)dx \approx \frac{(b-a)}{2n} \left[f(x_0) \sum_{i=1}^{n-1}2f(x_i) + f(x_n)\right]$$

Na expressão $a$ e $b$ são os limites de integração e $n$ é o número de intervalos

Vamos considerar a mesma expressão da aula de interpolação $f(x) = \frac{1}{2}x^3 - 3x^2+2x$ e vamos integrar entre os pontos 1 e 5. A integral exata vale $ I= -22$.

```r
# Utilizado apenas para reduzir tamanho dos gráficos
require(repr)
options(repr.plot.width=7, repr.plot.height=5)

# Pontos exatos da função proposta

funExata <- function(x){
    y <- 0.5*x^3 - 3*x^2 + 2*x
    return(y)
}

#funExata <- function(x){
#    y <- log(1/x) +0.1*x^2
#    return(y)
#}

# Plotando função e pontos que serão considerados para fazer interpolação e o cálculo da integral
xx <- seq(0,5.4,by=0.02)
plot(xx,funExata(xx), t='l', xlab='x', ylab='f(x)')
grid()
points(1:5, funExata(1:5), col=2, pch=19)
abline(h=0, lty=2)
```

```r
# Da aula de interpolação...
lagrange <- function(k, p, x){

    L <- 1
    np <- length(p)
    for(i in 1:np){
        if(i!=k){
            L <- L*(x-p[i])/(p[k]-p[i])
        }
    }    
    return(L)
}

solveLagrange <- function(x,y,dh,grau){
    nh <- length(x)-1
    ni <- nh/grau
    data <- c()
    for(i in 1:ni){
        w <- 1 + (i-1)*grau
        k <- w+grau
        dx <- seq(x[w], x[k], by=dh)
        xx <- x[w:k]
        yy <- y[w:k]
        v <- c()
        for(j in 1:(grau+1)){
            vtemp <- lagrange(j,xx,dx)*yy[j]
            v <- cbind(v,vtemp)
        }
        dtemp <- cbind(dx,rowSums(v))
        data <- rbind(data,dtemp)
    }
    return(data)
}

# interpolação linear
# função unique elimina pontos repetidos
linear <- unique(solveLagrange(x=1:5,y=funExata(1:5), dh=1, grau=1))

xx <- seq(0,5.4,by=0.02)
plot(xx,funExata(xx), t='l', xlab='x', ylab='f(x)')
grid()
points(1:5, funExata(1:5), col=2, pch=19)
lines(linear[,1:2], col=2)
polygon(x=c(1,linear[,1],5,1), y=c(0,linear[,2],0,0), col = 2, density = 15)
abline(h=0, lty=2)
```

```r
# Função para o cálculo da integral pela regra do trapézio
# a e b: limites de integração
# fun: função considerada
# n: número de intervalos
intTrapezio <- function(a, b, fun, n){
    x <- seq(a,b,length.out=(n+1))
    nx <- length(x)
    
    Itemp <- fun(x[1])/2 + fun(x[nx])/2
    
    for(i in 2:n){
        Itemp <- Itemp + fun(x[i])
    }
    I <- (b-a) * Itemp/n
    return(I)
}
```

```r
# Calculando a integral para o caso considerado

I1 <- intTrapezio(1,5,funExata,4)
I1
```

```r
# Aumentando o numero de intervalos...
intTrapezio(1,5,funExata,100)
```

## Regra 1/3 de Simpson

A regra 1/3 de Simpson é obtida a partir da interpolação utilizando polinômios de segundo grau de Lagrage. Para dois intervalos (3 nós) a regra obtida é 

$$ \int_{x_0}^{x_2} f(x)dx \approx \frac{h}{3}[f(x_0) + 4f(x_1) + f(x_2)] $$

sendo que $h$ vale $(b-a)/n$.

Generalizando, assim como foi feito para a regra do trapézio, encontra-se a expressão abaixo. Da mesma forma que para a regra do trapézio, a expressão generalizada como segue funciona apenas para malha com espaçamento regular.

$$ \int_a^b f(x)dx \approx \frac{h}{3} \left[f(x_0)+f(x_m)+2\sum_{i=1}^{m/2-1}f(x_{2i}) + 4\sum_{i=1}^{m/2}f(x_{2i-1})\right] $$

```r
# interpolação segundo grau
# função unique elimina pontos repetidos
quad <- unique(solveLagrange(x=1:5,y=funExata(1:5), dh=0.05, grau=2))

xx <- seq(0,5.4,by=0.02)
plot(xx,funExata(xx), t='l', xlab='x', ylab='f(x)')
grid()
points(1:5, funExata(1:5), col=2, pch=19)
lines(quad[,1:2], col=2)
polygon(x=c(1,quad[,1],5,1), y=c(0,quad[,2],0,0), col = 2, density = 15)
abline(h=0, lty=2)
```

```r
# Função para o cálculo da integral pela regra 1/3 de Simpson
# a e b: limites de integração
# fun: função considerada
# n: número de intervalos
intSimpson1_3 <- function(a, b, fun, n){
    x <- seq(a,b,length.out=(n+1))
    nx <- length(x)
    idEven <- seq(2, (nx-1), by=2)

    Itemp <- fun(x[1])/3 + fun(x[nx])/3
    
    # indexação no R começa por 1
    
    if(nx>3){
        idOdd <- seq(3, (nx-1), by=2)
        for(i in idOdd){
            Itemp <- Itemp + (2/3)*fun(x[i])
        }
    }
    
    for(k in idEven){
        Itemp <- Itemp + (4/3)*fun(x[k])
    }
    I <- (b-a) * Itemp/n
    return(I)
}
```

```r
# Calculando a integral para o caso considerado

I2 <- intSimpson1_3(1,5,funExata,2)
I2
```
